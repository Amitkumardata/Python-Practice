# DAY 5 Theme: Time Travel & Complex Logic
* Real-world data almost always involves Dates. Bosses want to know "Sales by Month" or "Growth per Week." 
* Also, simple filtering isn't enough; sometimes you need to create your own logic (e.g., "If profit is negative, mark it as LOSS").

```python
import pandas as pd
import numpy as np

# A dataset representing sales across different regions
data = {
    'Date': ['2023-01-15', '2023-01-20', '2023-02-10', '2023-02-12', '2023-01-15', '2023-03-05'],
    'Region': ['North', 'South', 'North', 'East', 'South', 'West'],
    'Product': ['Widget', 'Widget', 'Gadget', 'Widget', 'Gadget', 'Gadget'],
    'Sales': [100, 200, 150, 300, 120, 400],
    'Profit': [20, -50, 30, 60, -10, 100] # Negative profit means a loss
}

df = pd.DataFrame(data)
print("Day 5 Data Loaded!")
print(df)
print(df.info()) # Pay attention to the Dtype of 'Date'
```

# Converting dates
* Convert the date column from an "OBJECT" (TEXT) To a "DATETIME" object
* "2023-01-15" is just text, like "Banana". It doesn't know that January comes before February. To analyze time, we must convert it.

```python
# Convert text to date time 
df['Date']= pd.to_datetime(df['Date'])
print(df['Date'])

# noticed 'Date' is now datetime64

print(df.info())
```

# The Monthly Report 
* Task: Create a new column called Month that contains the name of the month (e.g., "January", "February").


```python
df['month']= df['Date'].dt.month_name() 
print(df['month'])
# change into year

df['year']= df['Date'].dt.year
print(df['year'])
```

# The complex pivot (multi grouping)
* calculate the total sales by region AND Month

```python
# pass the list ['region','month'] groupby

region_monthly_sales=df.groupby(['Region','Date'])['Sales'].sum()
print(region_monthly_sales)
```

# The logic flag (conditional logic)
* Create a new column called performance 
* If profit is greater than 0 , call it "profitable"
* If profit is 0 or less , call it" loss"


```python
# Syntax np.where (condition,value_if_true,value_if_false)
# -50 row show loss

df['Performance']= np.where(df['Profit'] >0, "Profitable","Loss")
print(df)
```

# DAY 9 Outliers and Statistics.
* We have a dataset of employee salaries at a startup. Most people earn normal wages, but there is one massive outlier
* (probably the CEO or a data entry error).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a dataset with a clear outlier
data = {
    'Employee_ID': range(101, 111),
    'Department': ['IT', 'IT', 'HR', 'HR', 'Sales', 'Sales', 'IT', 'Sales', 'HR', 'Executive'],
    'Salary': [50000, 52000, 48000, 49000, 55000, 54000, 53000, 58000, 47000, 1000000] # Look at the last one!
}

df = pd.DataFrame(data)
print("Day 9 Data Loaded!")
```

# 1: use the single comnad for stats, look at the row starts in single frame this is the first thing I do in any new dataset it tell the shape f the data
* Describe() generate descriptive dtstaistics 



```python
stats=df['Salary'].describe()
print(stats)

# we usally format it to look nicer(2 decimal places)
print(stats.round(2))
# the mean is massive because of the outlier 
```

# 2 : Create the Boxplot to visually idenitfy the outlier 
* The box hold the 50 % of people
* the line in the middle is the median
* The Dots outside the "Whiskers" are the Outlier 

```python
plt.figure(figsize=(8,4))

# x='Salary' plots it horizontally

sns.boxplot(x=df['Salary'],color='Green')
plt.title('Salary distibution(spot the outlier)')
plt.show()
# see the right side Dot way over 1000000
```

# 3: Calculate the 25th percentile(Q1) AND 74TH PERENTILE(Q3) to define the "Normal Range"
* We use the Interqourtile Range ( IQR Method)
*  IQR = The Gap between Q1 AND Q3

```python
Q1 =df['Salary'].quantile(0.25)
Q3 =df['Salary'].quantile(0.75)
IQR =Q3-Q1
print(f"Q1 (25%): {Q1}")
print(f"Q3 (75%): {Q3}")
print(f"IQR range: {IQR}")
```

#  4: The "Filter" (Removing Outliers)
* Task: Filter the dataframe to keep only the rows where the salary is less than Q3 + 1.5 * IQR. (This is the standard formula for "Normal").

```python
upper_limit =Q3+(1.5*IQR)
print(F" The math cutoff for an outlier is: {upper_limit}")

# create a 'clean' Dataframe
clean_df=df[df['Salary']<upper_limit]
print("n\--cleaned data --")
print(clean_df)
```

# 5: The Picture Histogram
* Task: Plot a hitogram of the Cleared data

```python
# kde =True adds a smooth curve line over the bars

sns.histplot(clean_df['Salary'], kde=True, color='Green')

plt.title('True Salary Distribution(without Outliers)')
plt.show()
```
